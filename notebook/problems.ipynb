## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, n, prob=0.5):
    '''Calculates the point estimate of the binomial distribution.'''

    return binom.pmf(k=x, n=n, p=prob, loc=0)

def pbinom(q, n, prob=0.5):
    '''Calculates the cumulative of the binomial distribution.'''

    return binom.cdf(k=q, n=n, p=prob, loc=0)

def qbinom(p, n, prob=0.5):
    '''Calculates the quantile function from the binomial distribution.'''

    return binom.ppf(q=p, n=n, p=prob, loc=0)

def rbinom(n, size, prob=0.5):
    '''Generates random variables from the binomial distribution.'''

    return binom.rvs(n=n, p=prob, size=size)
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, n=10, prob=0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, n=10, prob=0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, n=10, prob=0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(10, size=2000, prob=0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size=10
prob=0.2
data=rbinom(2000, size=10, prob=0.2)

distributions={
    'PMF': dbinom(data, size=size, prob=prob),
    'CDF': pbinom(data, size=size, prob=prob),
    'PPF': qbinom(data, size=size, prob=prob)
}

fig, axis=plt.subplots(3, 1, figsize=(5, 7))
axis=axis.flatten()

for i, distribution in enumerate(distributions.keys()):

    axis[i].set_title(f'{distribution}')
    axis[i].stem(data, distributions[distribution])
    axis[i].set_xlabel('Value')
    axis[i].set_ylabel('Probability')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
